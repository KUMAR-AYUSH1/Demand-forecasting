Feature creation

```python

df['Price_Premium_Index'] = df['Price'] / df['Competitor Pricing']
df['Discount_Depth'] = df['Discount'] / df['Price']

df['Date'] = pd.to_datetime(df['Date'])

df['Day_of_Week'] = df['Date'].dt.dayofweek
df['Is_Weekend']  = df['Date'].dt.dayofweek.isin([5, 6]).astype(int)
df['Month']       = df['Date'].dt.month
df['Day_of_Month']= df['Date'].dt.day
```

```python
df = df.sort_values(by=['Store ID', 'Product ID', 'Date']).reset_index(drop=True)
```

```python
df['demand_lag1'] = df.groupby(['Store ID', 'Product ID'])['Demand'].shift(1)
df['demand_lag2'] = df.groupby(['Store ID', 'Product ID'])['Demand'].shift(2)
df['demand_lag3'] = df.groupby(['Store ID', 'Product ID'])['Demand'].shift(3)
df['demand_lag4'] = df.groupby(['Store ID', 'Product ID'])['Demand'].shift(4)
df['demand_lag5'] = df.groupby(['Store ID', 'Product ID'])['Demand'].shift(5)
df['demand_lag6'] = df.groupby(['Store ID', 'Product ID'])['Demand'].shift(6)
df['demand_lag7'] = df.groupby(['Store ID', 'Product ID'])['Demand'].shift(7)
df['Units Sold_lag1'] = df.groupby(['Store ID', 'Product ID'])['Units Sold'].shift(1)
df['Units Sold_lag2'] = df.groupby(['Store ID', 'Product ID'])['Units Sold'].shift(2)
df['Units Sold_lag3'] = df.groupby(['Store ID', 'Product ID'])['Units Sold'].shift(3)
df['Units Sold_lag4'] = df.groupby(['Store ID', 'Product ID'])['Units Sold'].shift(4)
df['Units Sold_lag5'] = df.groupby(['Store ID', 'Product ID'])['Units Sold'].shift(5)
df['Units Sold_lag6'] = df.groupby(['Store ID', 'Product ID'])['Units Sold'].shift(6)
df['Units Sold_lag7'] = df.groupby(['Store ID', 'Product ID'])['Units Sold'].shift(7)

df['Units Ordered'] = df.groupby(['Store ID', 'Product ID'])['Units Ordered'].shift(1)

```

```python
df.dropna(inplace=True)
df.set_index('Date', inplace=True)
df.head()
```

```python
df.to_csv('preprocessed.csv')
```
